```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
wine = pd.read_csv('winery.csv', index_col = 0,low_memory=False)
display(wine)
```

```python
wine['designation'].unique()
```

```python
#First stepp will be to drop columns that don't add any values, in this case: region_2, taster_twitter_handle and designation. Designation has too many values and keeping variety will give as the needed informatrion about the type of wine without having to deal with too many values.
```

```python
wine = wine.drop(columns=['region_2', 'taster_twitter_handle', 'designation'])
wine.head()
```

```python
#the price is in dollar so we will convert it
wine['price'] = (wine['price']*0.93)
wine.head()
```

```python
wine.isnull().sum()
```

```python
#After looking at the missing values we will drop some columns and rows
#Rows: the one with the missing variety value, the ones with missing price because we want to be able to predict the price so we cannot substitute this value
#Columns: taster_name seems irrelevant to the prediction model as the magazine critics are not involved in the pricing process. Seeing as we have provinc as a subcategory of country we do not need to keep region_1 which has many missing values. We will try to fill in the missing country values based on other columns in the next steps.
#We will also drop title, as the information contained in the title is also contained in the columns variety and winery, the full name of each individual wine is far too specific to be usefull
wine = wine.drop(columns=['region_1', 'taster_name', 'title'])
wine.head()
```

```python
wine.dtypes
```

```python
wine = wine.dropna(subset=['price'])
```

```python
wine = wine.dropna(subset=['variety'])
```

```python
wine.isnull().sum()
```

```python
display(wine[wine['country'].isnull()])
```

```python
#categorise variety into red, white, rose, sparkling
wine['variety'].unique()
```

```python
#Blanc, garnacha, Red, Noir, Cabernet, Albariño, Rosado, Moscatel, Merlot, Sirah, White, white, Pedro Ximénez, Lambrusco, Shiraz, Malbec, Chardonnay, Muscat, Grenache, Rosé, Moscato, Blanco, Gris, Tinto, Malbec, Bianco, Merlot, Tempranillo, Rosa, Grigio, Clairette
```

```python
red_list = ['garnacha', 'red', 'noir', 'cabernet', 'merlot', 'sirah', 'shiraz', 'malbec', 'grenache', 'tinto', 'tempranillo', 'rouge']
white_list = ['blanc', 'albariño', 'moscatel', 'white', 'pedro ximénez', 'chardonnay', 'muscat', 'moscato', 'blanco', 'gris', 'bianco','grigio']
rose_list = ['rosado', 'lambrusco', 'rosé', 'rosa', 'clairette']

def identify_color(x):
    if any([(color in x.lower()) for color in red_list]):
        return 'red'
    elif any([(color in x.lower()) for color in white_list]):
        return 'white'
    elif any([(color in x.lower()) for color in rose_list]):
        return 'rose'
    else: 
        return 'unknown'
```

```python
wine['type'] =  wine['variety'].map(identify_color)
```
